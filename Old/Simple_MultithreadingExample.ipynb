```python
import threading

class DomainOperations:

    def __init__(self):
        self.FirstName = "Adrian"
        self.LastName = "Hood"

    def resolve_domain(self):
        self.domain_ip = 'hoodeventphoto.com'

    def generate_website_thumbnail(self):
        self.website_thumbnail= 'thumbnail_for_hoodeventphoto.jpg'

    def run(self):
        t1 = threading.Thread(target=self.resolve_domain)
        t2 = threading.Thread(target=self.generate_website_thumbnail)
        t1.start()
        t2.start()
        t1.join()
        t2.join()
        print(self.domain_ip, self.website_thumbnail)

```

```python
d = DomainOperations()
d.run()
```
